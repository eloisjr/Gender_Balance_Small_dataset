```python
df_real_values=df.copy()
df_real_values=df_real_values.rename(columns={'New South Wales DT PENALTY': 'NSW', 'Victoria DT PENALTY': 'VIC',
                                      'Queensland DT PENALTY': 'QLD', 'South Australia  DT PENALTY':'SA',
                                      'Western Australia DT PENALTY':'WA', 'Tasmania DT PENALTY': 'TAS',
                                      'Northern Territory DT PENALTY': 'NT','Australian Capital Territory DT PENALTY':'ACT'})


arima_pred2=arima_pred.rename(columns={'New South Wales DT PENALTY': 'NSW', 'Victoria DT PENALTY': 'VIC',
                                      'Queensland DT PENALTY': 'QLD', 'South Australia  DT PENALTY':'SA',
                                      'Western Australia DT PENALTY':'WA', 'Tasmania DT PENALTY': 'TAS',
                                      'Northern Territory DT PENALTY': 'NT','Australian Capital Territory DT PENALTY':'ACT'})


GM11_pred2=GM11_pred.rename(columns={'New South Wales DT PENALTY': 'NSW', 'Victoria DT PENALTY': 'VIC',
                                      'Queensland DT PENALTY': 'QLD', 'South Australia  DT PENALTY':'SA',
                                      'Western Australia DT PENALTY':'WA', 'Tasmania DT PENALTY': 'TAS',
                                      'Northern Territory DT PENALTY': 'NT','Australian Capital Territory DT PENALTY':'ACT'})


GM21_pred2=GM21_pred.rename(columns={'New South Wales DT PENALTY': 'NSW', 'Victoria DT PENALTY': 'VIC',
                                      'Queensland DT PENALTY': 'QLD', 'South Australia  DT PENALTY':'SA',
                                      'Western Australia DT PENALTY':'WA', 'Tasmania DT PENALTY': 'TAS',
                                      'Northern Territory DT PENALTY': 'NT','Australian Capital Territory DT PENALTY':'ACT'})
```

# ARIMA  VS GM11 COMPARISON


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the list of territories for the second dataset
territories_au = list(arima_pred2.columns.values)  # Replace df_2 with your second dataset

# Create subplots with 4 rows and 2 columns
fig, axs = plt.subplots(4, 2, figsize=(12, 16))

# Flatten the axs array for easier iteration
axs = axs.flatten()

# Loop through each territory
for i, state in enumerate(territories_au):
    # Plot real data from the first dataset
    axs[i].plot(df_real_values.index, df_real_values[state], label='Real values',color='darkblue',  marker='o')
    # Plot predictions from the first dataset
    axs[i].plot(GM11_pred2.index, GM11_pred2[state], label='GM11 predictions', color='orange', marker='o')
    # Set x-axis ticks and labels from 2004 to 2022 in steps of 2
    axs[i].set_xticks(ticks=np.arange(2004, 2023, 2))
    axs[i].set_xticklabels(labels=np.arange(2004, 2023, 2))

    # Plot real data from the second dataset
    axs[i].plot(arima_pred2.index, arima_pred2[state], label='ARIMA predictions', color='green',  marker='o')  # Adjust linestyle and color as needed
    # Plot predictions from the second dataset (if applicable)
    # axs[i].plot(pred_4_years_2.index, pred_4_years_2[state], label='Grey (1,1) Predictions (Dataset 2)', linestyle='dashdot', color='green')  # Adjust linestyle and color as needed

    # Set labels and title
    axs[i].set_xlabel('Years')
    axs[i].set_ylabel('Gender Balance index values')
    axs[i].set_title(f'Model Comparison: {state}')
    # Add legend
    axs[i].legend()
    # Hide grid
    axs[i].grid(False)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

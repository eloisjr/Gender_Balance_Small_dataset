# Open Metric Results

```python
arima_metrics=pd.read_csv("ARIMA_itertools_metrics.csv")

GM11_metrics=pd.read_csv("grey_model_11_metrics.csv")

GM21_metrics=pd.read_csv("metrics_GM21.csv")
```

# Compare Metrics

```python
arima_metrics_result=arima_metrics.T
arima_metrics_result=arima_metrics_result.reset_index(drop=True)
arima_metrics_result.columns = arima_metrics_result.iloc[0]
arima_metrics_result= arima_metrics_result[1:]
arima_metrics_result['index']=['NSW', 'VIC','QLD', 'SA', 'WA', 'TAS', 'NT', 'ACT']
arima_metrics_result =arima_metrics_result.set_index('index')

GM11_metrics_result= GM11_metrics.T
GM11_metrics_result = GM11_metrics_result.reset_index(drop=True)
GM11_metrics_result.columns = GM11_metrics_result.iloc[0]
GM11_metrics_result=GM11_metrics_result[1:]
GM11_metrics_result['index']=['NSW', 'VIC','QLD', 'SA', 'WA', 'TAS', 'NT', 'ACT']
GM11_metrics_result=GM11_metrics_result.set_index('index')

GM21_metrics_result= GM21_metrics.T
GM21_metrics_result = GM21_metrics_result.reset_index(drop=True)
GM21_metrics_result.columns = GM21_metrics_result.iloc[0]
GM21_metrics_result=GM21_metrics_result[1:]
GM21_metrics_result['index']=['NSW', 'VIC','QLD', 'SA', 'WA', 'TAS', 'NT', 'ACT']
GM21_metrics_result=GM21_metrics_result.set_index('index')

```

## MAPE

```python
metric_MAPE=pd.DataFrame()

metric_MAPE["Mape Arima"]=arima_metrics_result["MAPE"]
metric_MAPE["Mape GM11"]=GM11_metrics_result["MAPE"]
metric_MAPE["Mape GM21"]=GM21_metrics_result["MAPE"]
```

```python
#Calculate the median across all model and the lowest value indicates the best model.
metric_MAPE.median()
```

```python
metric_MAPE.sort_values('Mape GM21', inplace=False)

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(18, 6))  # Create a figure with two subplots side by side

# Plot the first subplot
metric_MAPE.plot(kind='bar', y='Mape GM21', ax=axs[0], color="Green")
axs[0].set_title('MAPE Results by State (GM21)')
axs[0].set_ylabel('Mape Values')
axs[0].set_xlabel('States/Territories')
axs[0].set_xticklabels(metric_MAPE.index, rotation=0, ha='right')
axs[0].grid(True)
axs[0].legend(loc='upper left')

# Sort the DataFrame for the second plot
metric_MAPE.sort_values('Mape GM11', inplace=True)

# Plot the second subplot
metric_MAPE[['Mape Arima', 'Mape GM11']].plot(kind='bar', ax=axs[1])
axs[1].set_title('MAPE Results by State (Arima and GM11)')
axs[1].set_ylabel('Mape Values')
axs[1].set_xlabel('States/Territories')
axs[1].set_xticklabels(metric_MAPE.index, rotation=0, ha='right')
axs[1].grid(True)
axs[1].legend(loc='upper left')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

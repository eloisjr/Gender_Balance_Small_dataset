# Open Male overseas Migration File

```python
# Rename columns for easy and compatibility find it.

########### Female Overseas Migration #############

male=pd.read_excel("EDA_MIGRATION.xlsx",'Male')
male['States']=['NSW', 'VIC','QLD', 'SA', 'WA', 'TAS', 'NT', 'ACT']
male = male.rename(columns={'States': 'State'})

########### Female Overseas Migration #############
female=pd.read_excel("EDA_MIGRATION.xlsx",'female')
female['State']=['NSW', 'VIC','QLD', 'SA', 'WA', 'TAS', 'NT', 'ACT']
```

# Open Labour Force File

```python
# Rename columns for easy and compatibility find it.

########### Male in labour force ############

male_lf=pd.read_excel("EDA_LABOUR.xlsx",'Male')
male_lf['States']=['NSW', 'VIC','QLD', 'SA', 'WA', 'TAS', 'NT', 'ACT']
male_lf= male_lf.rename(columns={'States': 'State'})


########### Female in labour force ###########

female_lf=pd.read_excel("EDA_LABOUR.xlsx",'Female')
female_lf['State']=['NSW', 'VIC','QLD', 'SA', 'WA', 'TAS', 'NT', 'ACT']
```

The scatter plot shows that The biggest cities such as NSW,QLD,SA, WA and Vic has a notable dropping in 2020, this is likely because COVID restrictions to work. Nevertheless, the rest of the states fluctuated a little from 2019 to 2021.

# MIGRATION: MALE & FEMALE


```python
states = male['State']
years = male.columns[1:]
cmap = get_cmap('twilight')

# Number of unique states
num_states = len(states)


######################  MALE OVERSEAS MIGRATION ######################
# Create figure and axes with a larger size
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))  # Adjust the size as needed

# Plot each state's data with colors on the first subplot
for i, state in enumerate(states):
    # Avoid red color
    color_index = (i + 1) / (num_states + 1)
    color = cmap(color_index)  # Adjusting for color range
    ax1.plot(years, male.loc[male['State'] == state, years].values.flatten(),
             label=state, color=color, marker='o')

# Add labels and title for the first subplot
ax1.set_xlabel('Year')
ax1.set_ylabel('Migration')
ax1.set_title('Overseas Migration: Male')

# Set x-axis range for the first subplot
ax1.set_xlim(2004, 2022)


###################### FEMALE OVERSEAS MIGRATION PLOT ######################

states2 = female['State']
years2 = female.columns[1:]
# Plot each state's data with colors on the second subplot
for i, state in enumerate(states2):
    color_index = (i + 1) / (num_states + 1)
    color = cmap(color_index)  # Adjusting for color range
    ax2.plot(years, female.loc[female['State'] == state, years2].values.flatten(),
             label=state, color=color, marker='o')

# Add labels and title for the second subplot
ax2.set_xlabel('Year')
ax2.set_ylabel('Migration')
ax2.set_title('Overseas Migration:Female')

# Place the legend on the right side outside of the plot for the second subplot
ax2.legend(title='States', bbox_to_anchor=(1.05, 1), loc='upper left')

# Set x-axis range for the second subplot
ax2.set_xlim(2004, 2022)

# Adjust layout
plt.tight_layout()

plt.savefig('migration_plot_EDA.png')
# Save or display the plot
plt.show()
```

**Observations:** The scatter plot shows that The biggest cities such as NSW,QLD,SA, WA and Vic has a notable dropping in 2020, this is likely because COVID restrictions to work. Nevertheless, the rest of the states fluctuated a little from 2019 to 2021.

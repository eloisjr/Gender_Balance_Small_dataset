```python
df=pd.read_excel("data_exclu_20_21.xlsx")
# Rename columns for easy and compatibility find it.
df['States']=['NSW', 'VIC','QLD', 'SA', 'WA', 'TAS', 'NT', 'ACT']
df = df.rename(columns={'States': 'State'})
states = df['State']
years = df.columns[1:]
```

```python
#Select colors
cmap = get_cmap('twilight')

# Number of unique states
num_states = len(states)

# Create figure and axis with a larger size
fig, ax = plt.subplots(figsize=(10, 6))  # Adjust the size as needed

# Plot each state's data with colors
for i, state in enumerate(states):
    # Avoid red color
    color_index = (i + 1) / (num_states + 1)
    color = cmap(color_index)  # Adjusting for color range
    ax.plot(years, df.loc[df['State'] == state, years].values.flatten(),
            label=state, color=color, marker='o')

# Add labels and title
ax.set_xlabel('Years')
ax.set_ylabel('Gender Balance index values')
ax.set_title('Gender Balance Index')

# Place the legend on the right side outside of the plot
ax.legend(title='States', bbox_to_anchor=(1.05, 1), loc='upper left')

# Set x-axis range
ax.set_xlim(2004, 2022)

# Adjust layout
plt.tight_layout()

# Save or display the plot
plt.show()
```
